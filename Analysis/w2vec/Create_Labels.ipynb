```python
#Read in research reccomendations
import pandas as pd
data = pd.read_csv('/data/research_rows.csv')
data.head()
```

```python
#Transform df 
dates = []
ratings = []
tickers = []
for ind in data.index: 
    tickers.append(data.iloc[ind][0].split()[0])
    ratings.append(data['3/31/2017'][ind])
    ratings.append(data['6/30/2017'][ind]) 
    ratings.append(data['9/29/2017'][ind])
    ratings.append(data['12/29/2017'][ind])
    ratings.append(data['3/30/2018'][ind])
    ratings.append(data['6/29/2018'][ind])
    ratings.append(data['9/28/2018'][ind])
    ratings.append(data['12/31/2018'][ind])
    ratings.append(data['3/29/2019'][ind])
    ratings.append(data['6/28/2019'][ind])
    ratings.append(data['9/30/2019'][ind])
    ratings.append(data['12/31/2019'][ind])
    dates.append('3/31/2017')
    dates.append('6/30/2017')
    dates.append('9/29/2017')
    dates.append('12/29/2017')
    dates.append('3/30/2018')
    dates.append('6/29/2018')
    dates.append('9/28/2018')
    dates.append('12/31/2018')
    dates.append('3/29/2019')
    dates.append('6/28/2019')
    dates.append('9/30/2019')
    dates.append('12/31/2019')
```

```python
#Get list of tickets
tickers = [item for item in tickers for i in range(12)]   
```

```python
#Create df in format necessary for modeling
labels_dict ={'tickers':tickers,
              'date':dates,
             'ratings':ratings}

labels_df = pd.DataFrame(labels_dict)
labels_df
```
